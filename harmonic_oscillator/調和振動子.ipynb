# 調和振動子

Euler法・4次のRunge-Kutta法を用いてシミュレーションを行い、全エネルギーの時間変化を観察する。
さらに、シンプレクティック法についても振る舞いを観察する。

なお、物体の運動方程式は次で与えられる。
$$ m\frac{d^2 x}{dt^2}=-kx $$

各定数は以下の通り。

```julia
const k=2.0
const m=1.0
const x₀=10.0
const v₀=0.0
```

$y_1:=x,\ y_2:=v (=\dot{x})$とすると、運動方程式は次のように書き換えられる。
$$ \frac{dy_1}{dt}=y_2 \\ \frac{dy_2}{dt}=-\frac{k}{m}y_1 $$
よって、$y:=(y1,y2)$としたとき$\dot{y}=f(x,y)$となるような$f$が下のセルのように書ける。

```julia
f(t,y) = [y[2],-k*y[1]/m]
```

エネルギーの様子をプロットする作業は共通なので、先に書いておく。

```julia
function plot_energy(method;isRough=false)
    Δt=isRough?1e-3:1e-5
    N=Int(ceil(30.0/Δt))
    y₀=[x₀, v₀]
    y=Array{Float64,2}(N+1,2)
    
    t=[i*Δt for i in 0:N]
    y[1,:]=y₀
    for i in 2:N+1
        y[i,:]=method(Δt*(i-1),y[i-1,:],Δt)
    end
        
    x=y[:,1]
    v=y[:,2]
    E=0.5m*v.^2+0.5k*x.^2
    E₀=0.5m*v₀^2+0.5k*x₀^2
    if isRough
        return Plots.plot(t,E-E₀,label="E-E_0")
    else
        return Plots.plot([t[i] for i in 1:100:N+1],[E[i]-E₀ for i in 1:100:N+1],label="E-E_0")
    end
end
```

## オイラー法

$h$を微小量とすると
$$ y_{n+1}-y_{n}=hf(t_n,y_n) $$
と書ける。

```julia
function next_euler(t,y,h=1e-5)
    return y+h.*f(t,y)
end
```

(陽的)オイラー法なのでﾓﾘﾓﾘエネルギーが大きくなっていく。

```julia
plot_energy(next_euler)
```

## 4次のルンゲクッタ法

```julia
function next_rk4(t,y,h=1e-5)
    k₁ = h*f(t, y)
    k₂ = h*f(t+h/2, y+k₁/2)
    k₃ = h*f(t+h/2, y+k₂/2)
    k₄ = h*f(t+h, y+k₃)
    
    return y+(k₁+2k₂+2k₃+k₄)/6
end
```

```julia
plot_energy(next_rk4)
```

不規則に見えるが、縦軸のスケールをみるとほぼ0であることが分かる。
以下のように$t=10^{-3}$まで刻み幅を粗くしても大差ない。

```julia
plot_energy(next_rk4,isRough=true)
```

## シンプレクティック法

```julia
function next_simplectic(t,y,h=1e-5)
    q=√k*y[1]
    p=√m*y[2]
    return [1 h; -h 1-h^2]*[q, p].*[1/√k; 1/√m]
end
```

```julia
plot_energy(next_simplectic)
```

エネルギーは一定の$O(h)$の範囲で保存する。
